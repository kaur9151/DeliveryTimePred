```python
import pandas as pd
```

```python
df=pd.read_csv('EDA_done.csv')
```

```python
df.head()
```

```python
df=df.iloc[:,1:]
```

```python
df=df.drop(labels=['Delivery_person_ID'],axis=1)
```

```python
df.head()
```

```python
#Independent and dependent features
X=df.drop(labels=['Time_taken (min)'],axis=1)
Y=df[['Time_taken (min)']]
```

```python
#Define which columns should be ordinal-encoded and which should be scaled
categorical_cols=X.select_dtypes(include='object').columns
numerical_cols=X.select_dtypes(exclude='object').columns
```

```python
# Define the custom ranking for each ordinal variable
Weather_map=['Stormy','Sandstorms','windy','Fog','Cloudy','Sunny']
traffic_map=['Jam','High','Medium','Low']
vehicle_map=['electric_scooter','scooter','motorcycle']
order_map=['Drinks','Snack','Meal','Buffet']
festival_map=['No','Yes']
city_map=['Semi-urban','urban','Metropolitian']
```

```python
from sklearn.impute import SimpleImputer #Handling Missing Values
from sklearn.preprocessing import StandardScaler #Handling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder #Ordinal Encoding
##pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
#Numerical Pipeline
num_pipeline=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='median')),
        ('scalar',StandardScaler())
    ]
)
#Categorical Pipeline
cat_pipeline=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('ordinalencoder',OrdinalEncoder(categories=[Weather_map,traffic_map,vehicle_map,order_map,festival_map,city_map])),
        ('scalar',StandardScaler())
    ]
)

preprocessor=ColumnTransformer([
    ('num_pipeline',num_pipeline,numerical_cols),
    ('cat_pipeline',cat_pipeline,categorical_cols)
])
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=30)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

```python
#Model Training
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae=mean_absolute_error(true,predicted)
    mse=mean_squared_error(true,predicted)
    rmse=np.sqrt(mean_squared_error(true,predicted))
    r2_square=r2_score(true,predicted)
    return mae,rmse,r2_square
```

```python
#Train multiple models
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}
trained_model_list=[]
model_list=[]
r2_list=[]
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)

    #Make Predictions
    y_pred=model.predict(X_test)

    mae,rmse,r2_square=evaluate_model(y_test,y_pred)
    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print('RMSE:',rmse)
    print('MAE:',mae)
    print('R2 score',r2_square*100)

    r2_list.append(r2_square)
    print('='*35)
    print('\n')
```
